````python
def write_vtt_from_captions(captions, filename):
    with open(filename,'w') as f:
        f.write('WEBVTT\n')
        for cap in captions:
            one_line = f"\n{cap['start']} --> {cap['end']} position:{cap['x']}% line:{cap['y']}% align:left\n"
            two_line = f"{cap['text']}\n"
            f.write(one_line)
            f.write(two_line)
````
