```python
def merge_lines_bbox(flines, index):
    word_in_frame_bbox = {}
    for k in flines.keys():
        bbox = [
            flines[k][0]['boundingBox'][0], # Left-Top
            flines[k][-1]['boundingBox'][1], # Right-Top
            flines[k][-1]['boundingBox'][2], # Right-Bottom
            flines[k][0]['boundingBox'][3], # Left-Bottom
        ]
        text = ' '.join([word['text'] for word in flines[k]])
        word_in_frame_bbox[xy_to_per(tuple(bbox[0]))] = {"bbox": bbox, "text":text, 'tidx': index}
    return word_in_frame_bbox
```

```python
def generate_vtt_lines(idx, cap):
    z_line = f"{idx}\n"
    one_line = f"{cap['start']} --> {cap['end']} position:{cap['x']}% line:{cap['y']}% align:left\n"
    two_line = f"{cap['text']}\n"
    return z_line, one_line, two_line
    

def write_vtt_from_captions(captions, filename):
    with open(filename,'w') as f:
        f.write('WEBVTT\n')
        for idx, cap in enumerate(captions):
            z_line, one_line, two_line = generate_vtt_lines(idx+1, cap)
            f.write('\n'+z_line)
            f.write(one_line)
            f.write(two_line)
```

```python

parse_int_from_pos = lambda x: int(x.split(':')[-1].replace('%',''))

def modify_vtt_from_parms(cue_id, action, params, edit_value):
    with open('gen_06.vtt', 'r') as f:
        s=f.readlines()
    target_index = s.index(f'{cue_id}\n')
    
    if action == "DELETE":
        del s[target_index-1:target_index+3]
    else:        
        target_cue = s[target_index:target_index+3]
        start_time, _, end_time, position, line, align= target_cue[1].split()

        target_dict = {
            "x": parse_int_from_pos(position),
            "y": parse_int_from_pos(line),
            "start": start_time,
            "end": end_time,
            "text": target_cue[2].strip()
        }
        target_dict[params] = edit_value
        s[target_index:target_index+3] =generate_vtt_lines(cue_id, target_dict)
    
    with open('gen_06_edit.vtt', 'w') as f:
        f.writelines(s)

modify_vtt_from_parms(1, 'EDIT', 'text', '수정테스트')
```
